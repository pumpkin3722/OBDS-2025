```python
import argparse


with open('gut_dnase1_chr21.bed', 'r') as beda,open('brain_dnase1_chr21.bed', 'r') as bedb:
    la=[]
    lb=[]
    for line in beda:
        result = line.split("\t")
        chra=result[0]
        starta=int(result[1])
        enda=int(result[2])
        la.append((chra, starta, enda))
        
    for line in bedb:
        result = line.split("\t")
        chrb=result[0]
        startb=int(result[1])
        endb=int(result[2])
        lb.append((chrb, startb, endb))
    
    interval=[]
    for inta in la:
        for intb in lb:
            if inta[0] == intb[0]:
                if intb[1] <= inta[2] and intb[2]>=inta[1]:
                    intervalstart=max(inta[1],intb[1])
                    intervalend=min(inta[2],intb[2])
                    interval.append((inta[0],intervalstart,intervalend))
                    break
    
interval                   
                
```
