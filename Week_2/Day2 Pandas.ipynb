```python
genes = ens_df[ens_df['feature']=='gene']
genes
```

Create a new column called “gene_length” that contains the length of each gene in dataframe genes.

```python
genes = genes.assign(gene_length = genes["end"]-genes["start"]+1)
genes.head(10)
```

Print the mean and median of the gene lengths.

```python
print(f'Mean gene length:{genes["gene_length"].mean()}')
print(f'Median gene length:{genes["gene_length"].median()}')
```

Find the top 10 longest genes and print their gene identifier.

```python
genes = genes.sort_values(by=["gene_length"], ascending =False)
top10=genes.head(10)
top10['attribute'].str[9:24]
```

```python
genes.nlargest(
    n = 10,
    columns = "gene_length"
)["attribute"].str.extract(
    pat = r'.*gene_id "([A-Za-z0-9]+)".*')
```

Find the lowest 20 shortest genes and print their gene identifier.

```python
genes = genes.sort_values(by=["gene_length"])
low20 = genes.head(20)
low20['attribute'].str[9:24]
```

Split the “attribute” column of the dataframe `genes` using `;` as delimiter, store the result in a new dataframe called `attributes`, then concatenante `genes` and `attributes`

```python
attributes = pd.DataFrame(genes['attribute'].str.split(";",expand=True))
attributes.columns=['gene_id','gene_version','gene_source','gene_biotype']
attributes['gene_id']=attributes['gene_id'].str[9:-1]
attributes['gene_version']=attributes['gene_version'].str[15:-1]
attributes['gene_source']=attributes['gene_source'].str[14:-1]
attributes['gene_biotype']=attributes['gene_biotype'].str[15:-1]
attributes

```

```python
pd.concat([genes, attributes], axis=1)
```
