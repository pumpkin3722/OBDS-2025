```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Exercise 1

Load read count dataset (featurecounts.tsv).

```python
featurecounts = pd.read_table("/project/shared/python/2_python_data_science/data/featurecounts.tsv", comment="#")
featurecounts
```

Remove unwanted columns (Chr, Start, End, Strand, Length) and rename the remaining columns.

```python
featurecounts.drop(columns=['Chr', 'Start', 'End', 'Strand', 'Length'], inplace=True)
featurecounts.columns = featurecounts.columns.str.replace(r'bam/([A-Za-z0-9-]+).bam', r'\1', regex=True)
featurecounts
```

Filter out genes with 0 counts across all samples.

```python
featurecounts = featurecounts[featurecounts.drop(columns = "Geneid").sum(axis=1) > 0]
featurecounts
```

Plot scatter plot of pairs of replicates.

```python
g = sns.PairGrid(featurecounts.drop(columns = "Geneid"))
g.map(plt.scatter)
```

Reshape the data in tidy format.

```python
featurecounts = featurecounts.melt(id_vars="Geneid", var_name="sample", value_name="counts")
featurecounts
```

Plot a histogram of read counts across all samples, on a log scale.

```python
sns.displot(data=featurecounts, x="counts", log_scale=True)
```

Plot a violin plot of read counts per sample.

```python
sns.violinplot(data=featurecounts, x="sample", y="counts")
```

Plot the count distribution of the 10 genes with highest expression (overall) in each sample.BONUS: rotate the x axis labels (i.e., gene names) 90 degrees.

```python
top_genes = featurecounts.groupby("Geneid")["counts"].sum().sort_values(ascending=False).head(10).index.to_list()
g = sns.FacetGrid(featurecounts[featurecounts['Geneid'].isin(top_genes)], col="sample", col_wrap=3)
g.map(sns.barplot, "Geneid", "counts")
# Rotate the x-axis labels
for ax in g.axes.flat:
    ax.tick_params(axis='x', rotation=90)
```

# Exercise 2

Load DeSeq2 results file (`deseq_output.tsv`)

```python
deseq = pd.read_table("/project/shared/python/2_python_data_science/data/deseq_output_kevin.tsv", index_col=0)
deseq
```

Prepare the dataframe for plotting
- You will need to transpose the table, but first:
- Make sure that future column names are stored in the index.
- The first column is not a gene name, remove it.

```python
deseq.drop(columns=deseq.columns[0], inplace=True)
deseq
```

```python
deseq = deseq.T
deseq
```

Plot a volcano plot of the results:
- X-axis: log2 fold change
- Y-axis: -log10 p-value
- Color points by adjusted p-value (FDR) > 0.05 (grey) or < 0.05 (red)
- Add a horizontal line at y = -log10(0.05) (dashed line)
- Add a vertical line at x = 0 (dashed line)


```python
deseq = deseq.assign(
    mlog10_pvalue = -np.log10(deseq["pvalue"])
).assign(de = deseq["padj"] < 0.05)
g = sns.relplot(
    deseq,
    x="log2FoldChange",
    y="mlog10_pvalue",
    hue="de",
    palette=sns.color_palette(["red", "grey"], 2),
    s = 1
)
g.axes[0][0].axhline(-np.log(0.05), ls='--')
g.axes[0][0].axvline(0, ls='--')
g._legend.set_title("padj < 0.05")
```

```python
deseq.drop(columns=["mlog10_pvalue", "de"]).to_csv("/project/shared/python/2_python_data_science/data/deseq_output_kevin.tsv", sep="\t")
```
