# Exercise 4

Create a new DataFrame called `genes` by subsetting all rows with “gene” feature type in `ens_df`.

```python
mask = ens_df["feature"] == "gene"
genes = ens_df[mask]
genes.shape
```

Create a new column called “gene_length” that contains the length of each gene in dataframe genes.

```python
genes = genes.assign(gene_length = genes["end"] - genes["start"] + 1)
genes.head(2)
```

Calculate the mean and median of the gene lengths.

```python
print("Mean gene length:", genes["gene_length"].mean())
print("Median gene length:", genes["gene_length"].median())
```

Find the top 10 longest genes and print their gene identifier.

```python
genes.nlargest(
    n = 10,
    columns = "gene_length"
)["attribute"].str.extract(
    pat = r'.*gene_id "([A-Za-z0-9]+)".*')
```

Find the lowest 20 shortest genes and print their gene identifier.

```python
genes.nsmallest(
    n = 20,
    columns = "gene_length"
)["attribute"].str.extract(
    pat=r'.*gene_id "([A-Za-z0-9]+)".*')
```

Split the “attribute” column of the dataframe `genes` using ';' as delimiter, store the result in a new dataframe called `attributes`, then concatenante `genes` and `attributes`

```python
attributes = genes["attribute"].str.split(
    ";", expand = True
)
attributes.columns = [
    'gene_id',
    'gene_version',
    'gene_source',
    'gene_biotype'
]
attributes.head()
```

```python
attributes['gene_id'] = attributes['gene_id'].str.replace('gene_id ','').str.replace('"','')
attributes['gene_version'] = attributes['gene_version'].str.replace('gene_version ','').str.replace('"','')
attributes['gene_source'] = attributes['gene_source'].str.replace('gene_source ','').str.replace('"','')
attributes['gene_biotype'] = attributes['gene_biotype'].str.replace('gene_biotype ','').str.replace('"','')
attributes.head()
```

```python
expanded = pd.concat([genes, attributes], axis=1)
expanded.head()
```

Using `genes`, count the number of genes per chromosome and add the number of genes per chomosome into the `idx_df` from part 1 using the `merge` method. NOTE: use `astype()` and `pd.StringDtype()` to convert chromosome names to `string` type.

```python
# Convert chromosome names to string type in both data sets
idx_df = idx_df.astype({'name': pd.StringDtype()})
genes = genes.astype({'seqname': pd.StringDtype()})
# Count the number of genes per chromosome
chr_gene_count = pd.DataFrame(genes["seqname"].value_counts())
# Add that information to idx_df
idx_df.merge(
    chr_gene_count,
    left_on="name",
    right_index=True)
```
